```python
import numpy as np
from scipy.stats import binom
```

```python
M = 10
y = np.array([
    7, 7, 1, 3, 2, 1, 5, 1, 4, 0, 
    2, 3, 3, 1, 2, 2, 3, 1, 4, 0, 
    8, 3, 5, 3, 0, 0, 7, 1, 1, 3, 
    1, 3, 2, 4, 1, 6, 2, 2, 4, 1, 
    3, 1, 1, 2, 7, 3, 3, 2, 2, 2
])
```

```python
np.random.seed(42)

pi = np.array([0.5, 0.5])
mu = np.random.rand(2)

pi1, mu1 = None, None

eps = 1e-7
```

```python
epoch = 0

while True:
    print(f'iteraton #{epoch}:')
    print(f'pi = {pi}')
    print(f'mu = {mu}')
    print()
    
    epoch += 1
    
    r = binom.pmf(y[:, None], M, mu[None, :]) * pi[None, :]
    r /= np.sum(r, axis=1)[:, None]
    
    pi1, mu1 = pi.copy(), mu.copy()
    
    pi = r.mean(axis=0)
    mu = np.sum(y[:, None] * r, axis=0) / r.sum(axis=0) / M
    
    if np.abs(pi - pi1).sum() < eps and np.abs(mu - mu1).sum() < eps:
        break
```

```python
print(f'pi = {pi}')
print(f'mu = {mu}')
```
